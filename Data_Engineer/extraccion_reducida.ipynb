```python
import tweepy
import pandas as pd
```

```python
# Tokens de acceso
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```python
# Autenticaci√≥n en la API de Twitter
auth = tweepy.OAuth1UserHandler(consumer_key, consumer_secret, access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
# Define las palabras clave relevantes para buscar los tweets
palabras_clave = ["bitcoin", "BTC", "$BTC"]
```

```python
# Obt√©n tweets basados en las palabras clave
tweets = []
for palabra_clave in palabras_clave:
    for tweet in tweepy.Cursor(api.search_tweets, palabra_clave, count=100, tweet_mode='extended').items(1000):
        tweets.append(tweet)
```

```python
# Almacena los tweets en un DataFrame de Pandas con las columnas solicitadas
tweets_df = pd.DataFrame(columns=['timestamp', 'user_id', 'screen_name', 'full_name', 'text', 'hashtags', 'mentions', 'retweet_count', 'favorite_count', 'tweet_sentiment'])

for tweet in tweets:
    timestamp = tweet.created_at
    user_id = tweet.user.id_str
    screen_name = tweet.user.screen_name
    full_name = tweet.user.name
    text = tweet.full_text
    hashtags = [tag['text'] for tag in tweet.entities['hashtags']]
    mentions = [mention['screen_name'] for mention in tweet.entities['user_mentions']]
    retweet_count = tweet.retweet_count
    favorite_count = tweet.favorite_count
    # Agrega la informaci√≥n de las columnas al DataFrame
    tweets_df = tweets_df.append({'timestamp': timestamp, 
                                  'user_id': user_id, 
                                  'screen_name': screen_name, 
                                  'full_name': full_name, 
                                  'text': text, 
                                  'hashtags': hashtags, 
                                  'mentions': mentions, 
                                  'retweet_count': retweet_count, 
                                  'favorite_count': favorite_count}, ignore_index=True)
```

```python
# Ordena los tweets por fecha y resetea el √≠ndice
tweets_df = tweets_df.sort_values(by='timestamp').reset_index(drop=True)
```
